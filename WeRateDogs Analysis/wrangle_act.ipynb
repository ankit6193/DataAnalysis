#### Quality Issues

- Dataset contains retweets 
- 'timestamp' and 'retweeted_status_timestamp' is of type String
- 'tweet_id', 'in_reply_to_status_id' , 'in_reply_to_user_id', retweeted_status_id, retweeted_status_user_id is of type float
- Sources contains URL and doesnot give clear indication of source
- Text contains numerator which does not match rating numerator column 
- 'rating numerator' and 'rating denominator' is of int datatype while some dogs have decimal in numerator hence need to change in to float
- name field has dummy names like 'a' , 'an' , 'the' 
- name field has value 'None' instead of NA
- Text contains value '&amp' which comes from API replacing '&' with '&amp' sometimes



 


#### Tidiness
- Dog Type ('doggo','floofer','pupper','puppo') does not need to be different columns since its specific
- Merging columns from data fetched from API and image prediction dataset to original dataset
